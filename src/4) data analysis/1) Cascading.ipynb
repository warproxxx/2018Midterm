```python
import pandas as pd
import swifter
import numpy as np
```

```python
df = pd.read_csv('../../data/all_cleaned.csv')
```

```python
df = df[df['Time'] > 1540947600]
```

```python
df['cascade'] = df['in_response_to']
```

```python
df.loc[df[df['cascade'] == 0].index, 'cascade'] = df[df['cascade'] == 0].ID #set cascade as ID if cascade = 0
```

```python
len(df)
```

```python
vals = df['cascade'].value_counts().reset_index()
```

```python
df = df[df['cascade'].isin(vals[vals['cascade'] > 1]['index'])]
```

```python
len(df)
```

```python
df.to_csv('data/retweetcascade.csv', index=None)
```
